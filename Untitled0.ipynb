```python
# ============================================
# URF Drive Pipeline — robust one-shot
# 1) Mounts YOUR My Drive (with retries)
# 2) Reads SHARED folder via Service Account (resolves shortcuts)
# 3) Exports Google Docs → HTML; parses JSON/HTML/TXT
# 4) Writes DOCX + JSON + TXT to My Drive / URF-processed / processed / YYYY-MM /
# Notes: Skips folders; handles inline images in DOCX
# ============================================

# Deps
!pip -q install google-api-python-client google-auth google-auth-httplib2 google-auth-oauthlib
!pip -q install python-docx beautifulsoup4 lxml html2text tqdm requests pillow

# ---- Config (leave as-is) ----
CREDENTIALS_JSON = r"""
{
  "type": "service_account",
  "project_id": "data-access-for-chatgpt-5",
  "private_key_id": "2d9c6b9036f76d29336a32a37a1db3a97fb5ec80",
  "private_key": "-----BEGIN PRIVATE KEY-----\nMIIEvQIBADANBgkqhkiG9w0BAQEFAASCBKcwggSjAgEAAoIBAQDw5un9KDBMJXx5\n6OH0kOvAHAweXoWR1fknDxb3IUVcAd4vYOs9SyHs7JUhT+FUDirU5kNHy9XQ8gX3\nuS48PxlSPYryQaF7K6x+W5vry/je7JXQXynzbaqf3DO0i6wx4/8gxfNEsAVNlZq5\n46TWhhtSffgVu8ObZrAgz/51Ejg3fEYbNolw1uKSm0fb1Hi5r0Dj29jijP3Fk/qe\nQUg3/dJ8vz9VrqS9pGLVCuCxDctycjsGasxsbSrAE+Jrk6pLOPPrOYnj4dCCEkak\nmzWaYiFbcGH8zPGw8neronNk/+NpDaT7tfx+lO98lxOUn9eSYNbHvvkNeIe+MkdE\nNvRK4AZRAgMBAAECggEAavUDN6ijxPMx0Iiy5dyAJ2Psh4FNCYJ4pDxhZGjrvB2C\nMgP0fdrBXclurwMh1qmRn8P6ZEKelK4YMzQPUKVtTyPO8J9uHFs4M0PmNutxfDyL\niZ/u5ssuodppRIRs+Buz9y+cQkdqm1dMgrMNSUfyko4mWzPp9bHCilb+sajct9kD\nVZXvQoVeyZQPPxOHK6B5twxMq6Fsx9pYCSjVzNtHcksjhzf9i+77SpOGKvCzzkzT\nPvaIP40BsH3lRudu63eqE42lOMDgoVr+VziOWtJebMYR1Va8yL3K6Va9W0LOldCQ\njndLFIJWA9OFDqWVfz2EVYDuIQaTUxkjoyq2UeFhtwKBgQD8RYpSVB4/2NuZoru/\n+52QTDuk68HUnXAEXmjXVepvZvQDE/TWWEC7s3P0xHpfik5qTNVPy/o2P8jq3Upp\nei2+yAE09VCgt9uZIQCj8S82gSA40PqP0HaDnzHhNMTuVCINUeJfY2KjtdAmnzsM\nKjVhcq+OaCKx7PgsN4dPp6XTJwKBgQD0dlttz05yBTBIjDRbwUUx2jaf/tEIwpWA\nTIQHYU/SgcDj9zF98p+Ymc1/dC0hGyrM8HND3r236weN6B3Ipl5JaSGnCSmimhHe\nnkaeJFjWysWA2BhOzHEQfuwmiTY/U7gZTT8bemejWtRg+NY/fCZgd7V27+ImzAlW\nE7wQ5pHlxwKBgGnm54JYPbqf1tBQOHEHt8VzuzYool4XGYc7+tcrAhdyKM2B+y5g\nLgyd8Pe344doEDMfvgyZnJhZfobaDvroh3unipoowYg2psFJ0XMBNMi0vEodsJHe\n+WGX2uCnvmQufX0J+vYAStBe6trLzYX10iLuejRWEvqOeEBXrz4JKtvRAoGAethD\nJ5JGREvRMJOr9/PONrgGPS5KktFdWSSaMC85cnobmTLt3vGeO4wGVSbnSryhLG0k\nLxXuVk639BqRZPQda1x2Ob8D+h7cHfeBWxS+w+kPiKiOxoI7yuNM/ixzYy1J1QUZ\nUlCXZeJN2REScMKNcmqVeNgg2AguJOAz4HZNGJ8CgYEAv+wEZL/RJJvTqdCR4iBc\nSIfMSSbXz4suv9udn5mcnh/G3ql+H7fqrGNcks6L8jOsTMh9ywRRD4uwcN+GUocg\nU/VUY4zGW4Dr43wm3VdB9OJ3JMly/XVmieo1+uLjyYFvgyqDa0KPmFjuZ69QQchH\nNVHN0cumfGHrFn2Vdhywlp0=\n-----END PRIVATE KEY-----\n",
  "client_email": "urf-agent@data-access-for-chatgpt-5.iam.gserviceaccount.com",
  "client_id": "110795329929526073863",
  "auth_uri": "https://accounts.google.com/o/oauth2/auth",
  "token_uri": "https://oauth2.googleapis.com/token",
  "auth_provider_x509_cert_url": "https://www.googleapis.com/oauth2/v1/certs",
  "client_x509_cert_url": "https://www.googleapis.com/robot/v1/metadata/x509/urf-agent%40data-access-for-chatgpt-5.iam.gserviceaccount.com",
  "universe_domain": "googleapis.com"
}
""".strip()

FOLDER_ID     = "14phps2mthRq3zkXdVg6VmTsWrdgY3mff"      # shared source folder (READ)
WRITE_JSON    = True
WRITE_TXT     = True
LOCAL_OUT     = "./out"                                   # temp workspace
MYDRIVE_ROOT  = "/content/drive/MyDrive/URF-processed"    # where outputs land

YOUR_NAME     = "Mark Aldiss"
YOUR_ALIASES  = {"me","user","you","ma","mark","m. aldiss","m a","m.a.","mark a."}

# ---- Imports ----
import io, os, re, json, base64, tempfile, requests, shutil, time
from datetime import datetime
from pathlib import Path
from typing import List, Dict, Any, Optional, Tuple
from tqdm import tqdm

from google.oauth2 import service_account
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
from googleapiclient.errors import HttpError

from bs4 import BeautifulSoup
import html2text
from docx import Document
from docx.shared import Inches
from PIL import Image

def ensure_dir(p: Path): p.mkdir(parents=True, exist_ok=True)
def safe_slug(s, maxlen: int = 64) -> str:
    if not s: s = "untitled"
    else: s = str(s).strip() or "untitled"
    s = re.sub(r"[^a-zA-Z0-9._-]+", "_", s)
    return (s[:maxlen] or "untitled").rstrip("_")

def parse_possible_date(s: str) -> Optional[str]:
    for pat in [r"(20\d{2})[-_/](0[1-9]|1[0-2])", r"(0[1-9]|1[0-2])[-_/](20\d{2})"]:
        m = re.search(pat, s)
        if m: return f"{m.group(1)}-{m.group(2)}" if pat.startswith("(20") else f"{m.group(2)}-{m.group(1)}"
    return None

def normalise_speaker(name: str) -> str:
    if not name: return name
    n = re.sub(r"[^\w\s\.]", "", name.strip().lower())
    if n in YOUR_ALIASES or n in {"author","role"}: return YOUR_NAME
    return name.strip()

# ============ 1) Mount *your* My Drive (with retries) ============
from google.colab import drive as gdrive
for attempt in range(3):
    try:
        print(f"[*] Mounting My Drive (attempt {attempt+1}/3)...")
        gdrive.mount('/content/drive', force_remount=True)
        print("[✓] My Drive mounted.")
        break
    except Exception as e:
        print(f"[!] Mount failed: {e}")
        if attempt == 2:
            raise
        print("...retrying in 5s — keep this tab in front and allow the popup.")
        time.sleep(5)

# ============ 2) Service Account client (READ ONLY) ============
def drive_client(creds_json_str: str):
    scope = ["https://www.googleapis.com/auth/drive.readonly"]
    creds_dict = json.loads(creds_json_str)
    with open("/content/creds.json","w",encoding="utf-8") as f: json.dump(creds_dict,f)
    credentials = service_account.Credentials.from_service_account_file("/content/creds.json", scopes=scope)
    return build("drive", "v3", credentials=credentials)

def list_folder(service, folder_id: str):
    q = f"'{folder_id}' in parents and trashed=false"
    files, page_token = [], None
    while True:
        resp = service.files().list(
            q=q,
            fields="nextPageToken, files(id,name,mimeType,modifiedTime,shortcutDetails)",
            pageToken=page_token
        ).execute()
        files += resp.get("files", [])
        page_token = resp.get("nextPageToken")
        if not page_token: break
    # filter out folders entirely
    files = [f for f in files if f.get("mimeType") != "application/vnd.google-apps.folder"]
    return files

def resolve_shortcut(service, file_obj):
    if file_obj.get("mimeType") == "application/vnd.google-apps.shortcut":
        target = file_obj.get("shortcutDetails", {}).get("targetId")
        if target:
            return service.files().get(
                fileId=target,
                fields="id,name,mimeType,modifiedTime"
            ).execute()
    return file_obj

EXPORT_MAP = {
    "application/vnd.google-apps.document": "text/html",
    "application/vnd.google-apps.spreadsheet": "text/csv",
    "application/vnd.google-apps.presentation": "text/plain",
}

def download_any(service, file_obj, dest: Path):
    file_obj = resolve_shortcut(service, file_obj)
    fid  = file_obj["id"]; mime = file_obj.get("mimeType","")
    if mime in EXPORT_MAP:
        data = service.files().export(fileId=fid, mimeType=EXPORT_MAP[mime]).execute()
        ext = ".html" if EXPORT_MAP[mime] == "text/html" else (".csv" if EXPORT_MAP[mime] == "text/csv" else ".txt")
        dest = dest.with_suffix(ext)
        dest.write_bytes(data)
        return dest
    # binary
    request = service.files().get_media(fileId=fid)
    fh = io.BytesIO(); downloader = MediaIoBaseDownload(fh, request); done = False
    while not done:
        status, done = downloader.next_chunk()
    dest.write_bytes(fh.getvalue())
    return dest

# ============ Parsers / Writers ============
def guess_text_from_html(html: str) -> str:
    try:
        soup = BeautifulSoup(html, "lxml")
        for tag in soup(["script","style","nav","header","footer","aside"]): tag.decompose()
        return soup.get_text("\n")
    except Exception:
        try:
            h = html2text.HTML2Text(); h.ignore_links = False
            return h.handle(html)
        except Exception:
            return html

def parse_json_export(p: Path):
    data = json.loads(p.read_text(encoding="utf-8")); out = []
    if isinstance(data, dict) and "conversations" in data:
        for conv in data["conversations"]:
            title = conv.get("title") or f"conversation_{conv.get('id','')}"
            msgs=[]
            for m in conv.get("messages", []):
                msgs.append({"ts":m.get("create_time") or m.get("ts") or "",
                             "speaker": normalise_speaker(m.get("author") or m.get("role") or ""),
                             "text": m.get("content") or m.get("text") or "", "images":[]})
            out.append((title, {"source":p.name, "thread_id":conv.get("id",""), "message_count":len(msgs)}, msgs))
        return out
    if isinstance(data, list):
        if data and isinstance(data[0], dict) and ("text" in data[0] or "content" in data[0]):
            msgs=[{"ts":m.get("ts") or "","speaker":normalise_speaker(m.get("speaker") or m.get("role") or ""),
                   "text":m.get("text") or m.get("content") or "","images":[]} for m in data]
            return [(safe_slug(p.stem), {"source":p.name,"message_count":len(msgs)}, msgs)]
        else:
            out=[]
            for i, conv in enumerate(data):
                title = safe_slug(conv.get("title", f"{p.stem}_{i}")); msgs=[]
                for m in conv.get("messages", []):
                    msgs.append({"ts":m.get("ts") or "","speaker":normalise_speaker(m.get("speaker") or m.get("role") or ""),
                                 "text":m.get("text") or m.get("content") or "", "images":[]})
                out.append((title, {"source":p.name,"index":i,"message_count":len(msgs)}, msgs))
            return out
    msgs = data.get("messages", []) if isinstance(data, dict) else []
    if not msgs: msgs=[{"ts":"","speaker":"","text":json.dumps(data, ensure_ascii=False),"images":[]}]
    return [(safe_slug(p.stem), {"source":p.name,"message_count":len(msgs)}, msgs)]

def _download_image_to_temp(src: str):
    try:
        if src.startswith("data:image/"):
            header,b64 = src.split(",",1); ext = header.split("/")[1].split(";")[0]
            raw = base64.b64decode(b64); tmp = tempfile.NamedTemporaryFile(delete=False, suffix="."+ext)
            tmp.write(raw); tmp.flush(); tmp.close(); path = tmp.name
        elif src.startswith(("http://","https://")):
            r = requests.get(src, timeout=15); r.raise_for_status()
            ext="png"; ctype=r.headers.get("Content-Type","")
            if "jpeg" in ctype: ext="jpg"
            elif "webp" in ctype: ext="webp"
            tmp = tempfile.NamedTemporaryFile(delete=False, suffix="."+ext); tmp.write(r.content); tmp.flush(); tmp.close()
            path = tmp.name
            if ext=="webp":
                img = Image.open(path).convert("RGB"); png_path = path.rsplit(".",1)[0]+".png"; img.save(png_path,"PNG"); path = png_path
        else:
            return None
        return path
    except Exception:
        return None

def parse_html_export(p: Path):
    html = p.read_text(encoding="utf-8", errors="ignore")
    soup = BeautifulSoup(html, "lxml")
    for tag in soup(["script","style","nav","header","footer","aside"]): tag.decompose()
    candidates = soup.select('[data-testid*="turn"], [data-message-id], div.message, article, div[class*="message"], div[class*="turn"]')
    messages=[]
    if candidates:
        for block in candidates:
            speaker = (block.get("data-author") or block.get("data-role"))
            if not speaker:
                lab = block.find(attrs={"class": re.compile("author|sender|role", re.I)})
                speaker = lab.get_text(strip=True) if lab else None
            ts = None
            tsel = block.find(attrs={"class": re.compile("time|date|timestamp", re.I)})
            if tsel: ts = tsel.get_text(strip=True)
            for tag in block.find_all(["h1","h2","h3"]):
                if tag.get_text(strip=True).lower() in {"you","assistant","system"}: tag.decompose()
            text_parts=[]
            for el in block.find_all(["p","li","pre","code","blockquote","span","div"]):
                if el.find("img"): continue
                t = el.get_text(" ", strip=True)
                if t: text_parts.append(t)
            txt = "\n".join(text_parts)
            imgs=[im.get("src") for im in block.find_all("img") if im.get("src")]
            if txt or imgs or speaker:
                messages.append({"ts": ts or "","speaker": normalise_speaker(speaker or ""),"text": txt,"images": imgs})
    if not messages:
        txt = guess_text_from_html(html); messages=[{"ts":"","speaker":"","text":txt,"images":[]}]
    return [(safe_slug(p.stem), {"source":p.name,"message_count":len(messages)}, messages)]

def parse_text_like(p: Path):
    txt = p.read_text(encoding="utf-8", errors="ignore")
    parts = re.split(r"\n-{3,}\n", txt); outs=[]
    for i, chunk in enumerate(parts):
        chunk = chunk.strip()
        if not chunk: continue
        outs.append((f"{p.stem}_part_{i+1}", {"source":p.name,"part_index":i+1},
                     [{"ts":"","speaker":YOUR_NAME,"text":chunk,"images":[]}]))
    return outs or [(p.stem, {"source":p.name}, [{"ts":"","speaker":YOUR_NAME,"text":txt,"images":[]}])]

def write_docx(title: str, meta: Dict[str,Any], messages: List[Dict[str,Any]], out_path: Path):
    doc = Document()
    doc.add_heading(title, level=1)
    info = doc.add_paragraph(); info.add_run("Metadata:\n").bold = True
    for k,v in meta.items(): info.add_run(f"{k}: {v}\n")
    for m in messages:
        ts = m.get("ts") or ""; speaker = m.get("speaker") or ""; text = m.get("text") or ""; images = m.get("images") or []
        doc.add_paragraph("")
        rp = doc.add_paragraph(); rp.add_run(f"[{ts}] {speaker}").bold = True
        if text: doc.add_paragraph(text)
        for src in images:
            tmp = _download_image_to_temp(src)
            if not tmp: continue
            try: doc.add_picture(tmp, width=Inches(5.5))
            except Exception: doc.add_paragraph(f"[image: {src}]")
    out_path.parent.mkdir(parents=True, exist_ok=True); doc.save(str(out_path))

def write_txt(title: str, meta: Dict[str,Any], messages: List[Dict[str,Any]], out_path: Path):
    lines = [f"# {title}", "", "Metadata:"]
    for k,v in meta.items(): lines.append(f"{k}: {v}")
    lines.append("")
    for m in messages:
        ts = m.get("ts") or ""; speaker = m.get("speaker") or ""; text = (m.get("text") or "").strip()
        lines.append(f"[{ts}] {speaker}")
        if text: lines.append(text)
        if m.get("images"):
            for src in m["images"]:
                lines.append(f"[image: {src}]")
        lines.append("")
    out_path.parent.mkdir(parents=True, exist_ok=True)
    out_path.write_text("\n".join(lines), encoding="utf-8")

def month_bucket(meta: Dict[str,Any], messages: List[Dict[str,Any]], fname: str, drive_modified: Optional[str]) -> str:
    for m in messages:
        ts = m.get("ts") or ""
        for fmt in ("%Y-%m-%d","%Y-%m-%d %H:%M:%S","%Y-%m-%dT%H:%M:%S","%Y-%m-%dT%SZ"):
            try: return datetime.strptime(str(ts)[:len(fmt)], fmt).strftime("%Y-%m")
            except Exception: pass
    mb = parse_possible_date(fname)
    if mb: return mb
    if drive_modified:
        try: return datetime.fromisoformat(drive_modified.replace("Z","+00:00")).strftime("%Y-%m")
        except Exception: pass
    return "unknown"

# ============ 3) List & download/export sources ============
service = drive_client(CREDENTIALS_JSON)
print("[*] Listing shared folder…")
drive_files = list_folder(service, FOLDER_ID)
print(f"Found {len(drive_files)} items (folders skipped)")

cache_dir = Path(LOCAL_OUT)/"cache"; ensure_dir(cache_dir)
docs_dir  = Path(LOCAL_OUT)/"docx"; ensure_dir(docs_dir)
json_dir  = Path(LOCAL_OUT)/"json"; ensure_dir(json_dir)
txt_dir   = Path(LOCAL_OUT)/"txt";  ensure_dir(txt_dir)

for f in tqdm(drive_files, desc="Downloading/exporting"):
    try:
        dest = cache_dir / f["name"]
        if not dest.exists():
            _ = download_any(service, f, dest)
    except HttpError as e:
        print(f"[skip] {f.get('name')} ({f.get('id')}): {getattr(e, 'error_details', None) or str(e)}")
    except Exception as e:
        print(f"[skip] {f.get('name')} ({f.get('id')}): {e}")

# ============ 4) Process to DOCX/JSON/TXT ============
created: List[Tuple[Path,str]] = []
meta_by_name = {f["name"]: f for f in drive_files}

for p in sorted(cache_dir.iterdir()):
    suff = p.suffix.lower()
    if   suff == ".json":        convs = parse_json_export(p)
    elif suff in {".txt",".md"}: convs = parse_text_like(p)
    elif suff in {".html",".htm"}: convs = parse_html_export(p)
    else:                         continue
    mod = meta_by_name.get(p.name, {}).get("modifiedTime")
    for title, meta, msgs in convs:
        month = month_bucket(meta, msgs, p.name, mod)

        out_docx = docs_dir/month/(safe_slug(title)+".docx")
        write_docx(title, {"source": meta.get("source", p.name), "message_count": meta.get("message_count", len(msgs))}, msgs, out_docx)
        created.append((out_docx, month))

        if WRITE_JSON:
            out_json = json_dir/month/(safe_slug(title)+".json")
            out_json.parent.mkdir(parents=True, exist_ok=True)
            with open(out_json,"w",encoding="utf-8") as jf:
                json.dump({"title":title,"meta":meta,"messages":msgs}, jf, ensure_ascii=False, indent=2)
            created.append((out_json, month))

        if WRITE_TXT:
            out_txt = txt_dir/month/(safe_slug(title)+".txt")
            write_txt(title, {"source": meta.get("source", p.name), "message_count": meta.get("message_count", len(msgs))}, msgs, out_txt)
            created.append((out_txt, month))

print(f"[✓] Processed {len(created)} output files.")

# ============ 5) Mirror to YOUR My Drive ============
mydrive_root = Path(MYDRIVE_ROOT)
mirrored = 0
for local_path, month in tqdm(created, desc="Mirroring to My Drive"):
    rel_type = local_path.parents[2].name   # docx/json/txt
    dst = mydrive_root / "processed" / month / rel_type / local_path.name
    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy2(local_path, dst)
    mirrored += 1

print(f"[✓] Mirrored {mirrored} files to: {mydrive_root}/processed/<YYYY-MM>/(docx|json|txt)")
print("Now open Google Drive → My Drive → URF-processed → processed")
```

```python
# --- Ultra-safe XML sanitizer + rebuild pass ---

from pathlib import Path
import json, re, shutil, base64, tempfile, requests
from datetime import datetime, timezone
from docx import Document
from docx.shared import Inches
from PIL import Image

BASE = Path("./out")
CACHE = BASE/"cache"
DOCS  = BASE/"docx"
JSONO = BASE/"json"
TXTS  = BASE/"txt"
MYDRIVE_ROOT = Path("/content/drive/MyDrive/URF-processed")

# XML 1.0 allowed ranges: 0x9 | 0xA | 0xD | 0x20–0xD7FF | 0xE000–0xFFFD | 0x10000–0x10FFFF
def xml_safe(s: str) -> str:
    if s is None:
        return ""
    if not isinstance(s, str):
        s = str(s)
    out_chars = []
    for ch in s:
        cp = ord(ch)
        if (
            cp == 0x09 or cp == 0x0A or cp == 0x0D or
            (0x20 <= cp <= 0xD7FF) or
            (0xE000 <= cp <= 0xFFFD) or
            (0x10000 <= cp <= 0x10FFFF)
        ):
            # also strip lingering NULLs just in case
            if cp != 0x00:
                out_chars.append(ch)
        else:
            # replace illegal with a space to keep layout
            out_chars.append(" ")
    return "".join(out_chars)

def safe_slug(s, maxlen=64):
    s = xml_safe(s).strip() or "untitled"
    s = re.sub(r"[^a-zA-Z0-9._-]+", "_", s)
    return (s[:maxlen] or "untitled").rstrip("_")

def _download_image_to_temp(src: str):
    try:
        if src.startswith("data:image/"):
            header,b64 = src.split(",",1); ext = header.split("/")[1].split(";")[0]
            raw = base64.b64decode(b64); tmp = tempfile.NamedTemporaryFile(delete=False, suffix="."+ext)
            tmp.write(raw); tmp.flush(); tmp.close(); path = tmp.name
        elif src.startswith(("http://","https://")):
            r = requests.get(src, timeout=15); r.raise_for_status()
            ext="png"; ctype=r.headers.get("Content-Type","")
            if "jpeg" in ctype: ext="jpg"
            elif "webp" in ctype: ext="webp"
            tmp = tempfile.NamedTemporaryFile(delete=False, suffix="."+ext); tmp.write(r.content); tmp.flush(); tmp.close()
            path = tmp.name
            if ext=="webp":
                img = Image.open(path).convert("RGB"); png_path = path.rsplit(".",1)[0]+".png"; img.save(png_path,"PNG"); path = png_path
        else:
            return None
        return path
    except Exception:
        return None

def write_docx_safe(title, meta, messages, out_path: Path):
    title = xml_safe(title)
    doc = Document()
    doc.add_heading(title, level=1)

    p = doc.add_paragraph(); r = p.add_run("Metadata:\n"); r.bold = True
    for k,v in (meta or {}).items():
        k2 = xml_safe(k); v2 = xml_safe(v)
        doc.add_paragraph(f"{k2}: {v2}")

    for m in messages or []:
        ts      = xml_safe(m.get("ts") or "")
        speaker = xml_safe(m.get("speaker") or "")
        text    = xml_safe(m.get("text") or "")
        images  = m.get("images") or []
        doc.add_paragraph("")
        rp = doc.add_paragraph(); rp.add_run(f"[{ts}] {speaker}").bold = True
        if text:
            # If anything still slips through, wrap in try/except
            try:
                doc.add_paragraph(text)
            except Exception:
                doc.add_paragraph(xml_safe(text))
        for src in images:
            tmp = _download_image_to_temp(src)
            if not tmp:
                continue
            try:
                doc.add_picture(tmp, width=Inches(5.5))
            except Exception:
                doc.add_paragraph(f"[image: {xml_safe(src)}]")

    out_path.parent.mkdir(parents=True, exist_ok=True)
    doc.save(str(out_path))

def write_txt_safe(title, meta, messages, out_path: Path):
    lines = [f"# {xml_safe(title)}", "", "Metadata:"]
    for k,v in (meta or {}).items():
        lines.append(f"{xml_safe(k)}: {xml_safe(v)}")
    lines.append("")
    for m in messages or []:
        ts      = xml_safe(m.get("ts") or "")
        speaker = xml_safe(m.get("speaker") or "")
        text    = xml_safe(m.get("text") or "")
        lines.append(f"[{ts}] {speaker}")
        if text: lines.append(text)
        if m.get("images"):
            for src in m["images"]:
                lines.append(f"[image: {xml_safe(src)}]")
        lines.append("")
    out_path.parent.mkdir(parents=True, exist_ok=True)
    out_path.write_text("\n".join(lines), encoding="utf-8")

def month_from_epoch(ts) -> str:
    try:
        dt = datetime.fromtimestamp(float(ts), tz=timezone.utc)
        return dt.strftime("%Y-%m")
    except Exception:
        return "unknown"

# Re-load the mapping JSONs already in ./out/cache and rebuild safely
def parse_json_export_chatgpt(p: Path):
    data = json.loads(p.read_text(encoding="utf-8"))
    if isinstance(data, dict) and "conversations" in data: convs = data["conversations"]
    elif isinstance(data, dict) and "mapping" in data:     convs = [data]
    elif isinstance(data, list) and data and isinstance(data[0], dict) and "mapping" in data[0]: convs = data
    else: return []

    outs = []
    for conv in convs:
        title = conv.get("title") or conv.get("id") or p.stem
        mapping = conv.get("mapping", {})
        msgs = []
        for node in mapping.values():
            msg  = (node or {}).get("message") or {}
            role = ((msg.get("author") or {}).get("role") or "").strip()
            if role == "system":
                continue
            ts = msg.get("create_time") or node.get("create_time") or ""
            content = msg.get("content") or {}
            text = ""
            if isinstance(content, dict):
                parts = content.get("parts")
                if isinstance(parts, list):
                    text = "\n".join([str(x) for x in parts if x])
                elif isinstance(parts, str):
                    text = parts
            elif isinstance(content, str):
                text = content
            text = xml_safe(text)
            if text:
                msgs.append({
                    "ts": ts,
                    "speaker": "Mark Aldiss" if role == "user" else ("Assistant" if role else ""),
                    "text": text,
                    "images": []
                })
        # sort by epoch
        def _k(m):
            try: return float(m.get("ts") or 0)
            except: return 0.0
        msgs.sort(key=_k)
        outs.append((title, {"source": p.name, "message_count": len(msgs)}, msgs))
    return outs

created = []
for p in sorted(CACHE.iterdir()):
    if p.suffix.lower() != ".json":
        continue
    convs = parse_json_export_chatgpt(p)
    if not convs:
        continue
    for title, meta, msgs in convs:
        month = "unknown"
        for m in msgs:
            month = month_from_epoch(m.get("ts"))
            if month != "unknown": break
        out_docx = DOCS/month/(safe_slug(title)+".docx")
        write_docx_safe(title, meta, msgs, out_docx)
        out_txt  = TXTS/month/(safe_slug(title)+".txt")
        write_txt_safe(title, meta, msgs, out_txt)
        out_json = JSONO/month/(safe_slug(title)+".json")
        out_json.parent.mkdir(parents=True, exist_ok=True)
        out_json.write_text(json.dumps({"title":title,"meta":meta,"messages":msgs}, ensure_ascii=False, indent=2), encoding="utf-8")
        created.append((out_docx, month))

print(f"[✓] Rebuilt {len(created)} DOCX safely.")

# Mirror to Drive (overwrite)
mirrored = 0
for local_path, month in created:
    rel_type = local_path.parents[2].name
    dst = MYDRIVE_ROOT / "processed" / month / rel_type / local_path.name
    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy2(local_path, dst)
    mirrored += 1
print(f"[✓] Mirrored {mirrored} files to My Drive.")
```

```python
# ============================================
# URF Index v1 — Build a semantic search index from your Drive
# - Reads: My Drive / URF-processed / processed / <YYYY-MM> / (txt | json)
# - Embeds chunks with sentence-transformers
# - Stores: FAISS index + metadata in My Drive / URF-processed / index/
# - Delta mode: only (re)embeds files that changed
# ============================================

!pip -q install faiss-cpu sentence-transformers pandas tqdm orjson python-dateutil

import os, re, json, orjson, hashlib, math
from pathlib import Path
from datetime import datetime
from dateutil import tz
import pandas as pd
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

from google.colab import drive
drive.mount('/content/drive', force_remount=False)

# ----- Paths -----
ROOT = Path("/content/drive/MyDrive/URF-processed/processed")  # where we mirrored outputs
INDEX_DIR = Path("/content/drive/MyDrive/URF-processed/index") # where we'll save the index
INDEX_DIR.mkdir(parents=True, exist_ok=True)
LEDGER_PATH = INDEX_DIR / "ledger.json"        # tracks file versions so we can do delta
META_PATH   = INDEX_DIR / "metadata.parquet"   # rows describing each chunk
FAISS_PATH  = INDEX_DIR / "faiss.index"        # vector index
MODEL_NAME  = "sentence-transformers/all-MiniLM-L6-v2"  # small, fast, solid

# ----- Load / init model -----
model = SentenceTransformer(MODEL_NAME)
dim = model.get_sentence_embedding_dimension()

# ----- Helpers -----
def file_sig(p: Path) -> str:
    """A cheap content signature: size + mtime; enough for delta mode."""
    st = p.stat()
    return f"{st.st_size}-{int(st.st_mtime)}"

def iter_source_files(root: Path):
    for mdir in sorted(root.iterdir()):
        if not mdir.is_dir():
            continue
        for sub in ("txt","json"):  # prefer txt & json (skip docx for speed)
            d = mdir / sub
            if d.exists():
                for f in d.glob("*.txt"):
                    yield f
                for f in d.glob("*.json"):
                    yield f

def load_messages_from_txt(p: Path):
    # TXT format created earlier: lines like "[ts] Speaker" followed by text
    text = p.read_text(encoding="utf-8", errors="ignore")
    # Split by header-like markers; keep simple: treat whole file as one doc
    return [{"ts":"", "speaker":"", "text":text}]

def load_messages_from_json(p: Path):
    try:
        data = json.loads(p.read_text(encoding="utf-8"))
        msgs = data.get("messages", [])
        out = []
        for m in msgs:
            out.append({
                "ts": str(m.get("ts","")),
                "speaker": str(m.get("speaker","")),
                "text": str(m.get("text",""))
            })
        return out or [{"ts":"","speaker":"","text":""}]
    except Exception:
        return [{"ts":"","speaker":"","text":p.read_text(encoding="utf-8", errors="ignore")}]

CTRL_RE = re.compile(r"[\x00-\x08\x0B\x0C\x0E-\x1F]")
def clean_text(s: str) -> str:
    s = CTRL_RE.sub(" ", s or "")
    s = re.sub(r"\s+\n", "\n", s)
    s = re.sub(r"\n\s+", "\n", s)
    return s.strip()

def chunk_text(s: str, target_tokens=180, overlap=30):
    """
    Simple word-based chunking that roughly approximates tokens.
    """
    words = s.split()
    if not words:
        return []
    # convert to sizes
    step = max(1, target_tokens - overlap)
    chunks = []
    for start in range(0, len(words), step):
        piece = " ".join(words[start:start+target_tokens])
        if piece.strip():
            chunks.append(piece.strip())
    return chunks

def ts_to_iso(ts):
    # handle epoch seconds if present; else return as-is
    try:
        t = float(ts)
        dt = datetime.utcfromtimestamp(t)
        return dt.replace(tzinfo=tz.tzutc()).isoformat()
    except Exception:
        return str(ts)

# ----- Load ledger (delta) -----
if LEDGER_PATH.exists():
    try:
        ledger = json.loads(LEDGER_PATH.read_text())
    except Exception:
        ledger = {}
else:
    ledger = {}

new_rows = []
texts_to_embed = []

print("Scanning sources…")
for f in tqdm(list(iter_source_files(ROOT))):
    sig = file_sig(f)
    key = str(f)
    if ledger.get(key) == sig:
        continue  # unchanged; already indexed

    # load & normalize
    if f.suffix.lower() == ".txt":
        msgs = load_messages_from_txt(f)
    else:
        msgs = load_messages_from_json(f)

    # Build chunks
    for i, m in enumerate(msgs):
        base = clean_text(m.get("text",""))
        if not base:
            continue
        chunks = chunk_text(base, target_tokens=180, overlap=30)
        for j, ch in enumerate(chunks):
            texts_to_embed.append(ch)
            new_rows.append({
                "path": str(f),
                "file_name": f.name,
                "month": f.parent.parent.name,     # YYYY-MM
                "type": f.parent.name,             # txt/json
                "message_idx": i,
                "chunk_idx": j,
                "speaker": m.get("speaker",""),
                "ts_raw": m.get("ts",""),
                "ts_iso": ts_to_iso(m.get("ts","")),
                "text": ch
            })

    # mark as processed
    ledger[key] = sig

if not new_rows:
    print("No new or changed files. (Delta mode) Nothing to add.")
    # still ensure index + meta exist
    if META_PATH.exists() and FAISS_PATH.exists():
        print("Index already present at:", INDEX_DIR)
    else:
        print("No existing index found. Re-run after adding files.")
else:
    print(f"Preparing to embed {len(new_rows)} chunks…")

    # ---- Embed ----
    # Batch for memory
    batch = 256
    embs = []
    for i in tqdm(range(0, len(texts_to_embed), batch), desc="Embedding"):
        vecs = model.encode(texts_to_embed[i:i+batch], normalize_embeddings=True, convert_to_numpy=True)
        embs.append(vecs)
    X = np.vstack(embs).astype("float32")

    # ---- FAISS index (append or create) ----
    if FAISS_PATH.exists():
        index = faiss.read_index(str(FAISS_PATH))
        # sanity: dim match
        assert index.d == dim, f"Index dim {index.d} != model dim {dim}"
    else:
        index = faiss.IndexFlatIP(dim)  # cosine if vectors normalized

    index.add(X)
    faiss.write_index(index, str(FAISS_PATH))

    # ---- Append metadata ----
    df_new = pd.DataFrame(new_rows)
    if META_PATH.exists():
        df_old = pd.read_parquet(META_PATH)
        df_all = pd.concat([df_old, df_new], ignore_index=True)
    else:
        df_all = df_new
    df_all.to_parquet(META_PATH, index=False)

    # ---- Persist ledger ----
    LEDGER_PATH.write_text(json.dumps(ledger, indent=2))

    print(f"Done. Total vectors in index: {index.ntotal}")
    print("Saved to:", INDEX_DIR)

print("\nQuick usage notes:")
print("• Index: /MyDrive/URF-processed/index/faiss.index")
print("• Metadata: /MyDrive/URF-processed/index/metadata.parquet")
print("• Next runs only add new/changed files automatically.")
```
